在Python中，类型是在运行时自动决定的，而不是通过代码声明。  
  
a = 3的声明中会执行3个步骤：
1. 创建一个对象来代表值3
2. 创建一个变量a，若还没有创建的话
3. 将变量与新的对象3相连接
  
从变量到对象的连接称为引用，引用是一种关系，通过内存中的指针来实现，一旦使用变量，Python会自动跟踪这个变量到对象的连接
1. 变量是一个系统表的入口，包含了指向对象的连接
2. 对象是分配到的一块内存，有足够的空间去表示他们所代表的值
3. 引用是自动形成的从变量到对象的指针

```python
a = 1
print(a.__repr__)
print(a.__str__)
```

变量名没有类型，类型属于对象而不是变量名。  
每当一个变量名被赋予一个新的对象，如果原来的对象没有被其他的变量名或对象所引用的话，那个对象所占用的空间就会被回收，称为垃圾回收。  
在内部，Python在每个对象中保留了一个计数器，计数器记录当前指向该对象的引用的数目，一旦这个计数器被设置为零，这个对象的内存空间就会被自动回收。

# 共享引用
Python中不可能发生两个变量的相互关联。真实情况是两个变量通过他们的引用指向了同一个对象。  
Python中变量总是一个指向对象的指针，而不是可改变的内存区域的标签。  
对于支持在原位置修改的对象，共享引用时需要特别小心。

```python
L1 = [2, 3, 4]
L2 = L1
print(L1)
print(L2)
```

```python
L1[0] = 24
print(L1)
print(L2)
```

因列表这个对象是与其他对象共享的（被其他对象引用）。  
这种行为仅针对支持原位置改变的可变对象。  
若不想这样的现象发生，可以请求Python复制（copy）对象而不是创建引用。如：数组上使用从头到尾的分片、字典或集合上使用X.copy()。  

```python
L1 = [2, 3, 4]
L2 = L1[:]
L1[0] = 24
print(L1)
print(L2)
```

==运算符：测试两个被引用的对象是否有相同的值。  
is运算符：检查对象的同一性。
