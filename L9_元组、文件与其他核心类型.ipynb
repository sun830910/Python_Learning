# 存储Python原生对象：pickle
eval可以将字符串转换为对象。  
pickle模块可以让我们直接在文件存储几乎任何Python对象的高级工具，同时不需要对字符串来回转换。

```python
import pickle
D = {'a':1, 'b':2}
F = open('datafile.pkl', 'wb')  # wb为写入二进制
pickle.dump(D, F)
F.close()
```

```python
F = open('datafile.pkl', 'rb')
E = pickle.load(F)
E
```
