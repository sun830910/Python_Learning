# 第一章
## 为何使用Python?
1. 软件质量：Python更注重可读性、一致性和软件质量，具备比起传统脚本语言更优秀的可重用性和可维护性。
2. 开发者生产效率：相对于C、C++、Java等编译/静态类型语言，Python的开发者效率提升数倍，Python的代码长度往往只有C++或Java的1/5～1/3，意味着可以录入更少的代码，调试更少的代码，并在开发完成后维护更少的代码，并且可以立即运行，而无需传统编译/静态语言所必须的编译及链接等步骤，进一步提高了程序开发的效率。
3. 可移植性：大多数程序可以在不做任何改变即可在所有主流计算机平台上运行。
4. 标准库的支持：内置了众多预构建并可移植的功能模块。
5. 组件集成：可以通过多种集成机制轻松的与应用程序的其他部分进行通信。

## Python是一门脚本语言吗？
Python是一门通用型的编程语言，可以定义为一门面向对象的脚本语言，人们往往用“脚本”(script)而不是“程序”(program)一词来描述Python的代码文件。  
“脚本”往往倾向于描述简单的顶层代码文件，而“程序”则用来描述那些相对复杂一些的多文件应用。  

## Python的三个角色
1. Shell工具
2. 控制语言
3. 使用便捷

## Python的缺点
执行速度不够快。
目前Python的标准实现方式是将源代码的语句编译（翻译）为字节码（byte code）形式，之后再将字节码解释出来。  
字节码是一种与平台无关的格式，因而具有可移植性，然而，因为Python通常不会将代码编译成底层的二进制代码（例如Intel芯片的指令），一些Python程序将会比像C这样的完全编译语言慢一些。  

## Python的优点
1. 面向对象和函数式：Python是一种面向对象的语言，它的类模型支持多态、运算符重载和多重继承等高级概念。
2. 免费
3. 可移植：Python的标准实现是由可移植的ANSI C编写的，可以在目前所有主流平台上编译和运行。并且Python程序自动编译成可移植的字节码，这些字节码在已安装兼容版本Python的平台上运行结果都是相同的。
4. 功能强大：动态类型-Python在程序运行过程中跟踪对象的类型，不需要代码中进行关于复杂的类型和大小的声明。因为Python代码不约束数据类型，往往自动应用了一种广义上的对象。Python自动为对象分配内存，并且当对象不再使用时将自动撤销空间（垃圾回收），当需要自动扩展或收缩时，Python能够帮你完成底层的内存管理。
5. 可混合：Python可以以多种方式轻易的与其他语言编写的组件结合在一起。
6. 相对简单易用
7. 相对简单易学
8. 以Monty Python命名：BBC在1970年代的喜剧《Monty Python‘s Flying Circus》
