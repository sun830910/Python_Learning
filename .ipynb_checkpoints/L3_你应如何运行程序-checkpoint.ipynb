# 交互式命令行模式
## 开始一个交互式会话
最简单的运行方式，在Python交互式命令行中输入这些程序，有时候也称为交互式提示模式。  
可以在IDE、系统终端中等方法开始。  
在系统命令行下输入“python”后即可开始一个交互式等Python会话。

## 系统路径
当我们启动一个交互式对话并输入Python后，我们依赖于系统帮我们在程序搜索路径中定位Python程序等位置。  
根据Python版本和平台等不同，若没有设置系统的PATH环境变量来包括Python的安装路径，可能需要用你机器上的Python可执行文件的完整路径来代替单词“Python”。

# 系统命令行和文件
尽管交互命令行对于实验和测试来说都很好，但有一个很大的缺点：Python一旦执行了输入的程序之后就消失了，在交互式命令行下输入的代码是不会保存在一个文件中的。  
为了能够永久保存，需将代码写在文件中，这样的文件通常叫做模块。  
模块是一个包含了Python语句的简单文本文件。一旦编写完成，可以让Python解释器多次运行文件中的语句，并可以以多种方式进行运行。  


# 模块导入和重载
## 导入和重载基础知识
每一个以扩展名.py结尾的Python源文件都是一个模块，不需要任何特殊的代码或语法来使文件成为模块。  
在文件第一次导入模块后，其他的导入语句都不会再工作。因为导入是一个开销很大的操作，所以尽量避免重复导入。  
导入必须找到文件，将其编译成字节码，并且运行代码。  
若真的想要Python在同一文件中再次运行导入，则需要调用imp标准库中可用的reload函数（这个函数在Python2.X中为内置函数，但在Python3.X中不是内置函数）。  
reload函数自身载入并运行了文件当前版本但代码，若以及在另一个窗口中修改并保存了它，那将反映出修改变化。  
reload函数希望获得的参数是一个已经被加载了的模块对象名称，所以在重载之前，请确保已经成功导入了这模块。reload函数还希望在模块对象的名称周围有圆括弧，而import则不需要，reload是一个被调用的函数，而import是一个语句。
