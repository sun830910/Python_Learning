# break、continue、pass和循环的else
1. break：跳出最近所在的外围循环（跳过整个循环语句）
2. continue：跳到最近所在外围循环的头部（来到循环的头部行）
3. pass：什么事也不做，只是一条空占位语句
4. 循环else块：当且仅当循环正常离开时才会执行

# 并行遍历

```python
L1 = [1, 2, 3, 4]
L2 = [5, 6, 7, 8]
print(zip(L1, L2))
```

```python
list(zip(L1, L2))
```

```python
for (x, y) in zip(L1, L2):
    print(x , ' ', y, '==', x+y)
```
